```python
import seaborn as sns
# find out the mean price of the brand that have been purchased by users
price_list = df[df['event_type'] == 'purchase'].groupby('brand')['price'].max().sort_values(ascending=False)
```

```python
price_list[:20].plot(kind='bar')
```

```python
pltest = price_list.to_frame().iloc[:20].reset_index()
```

```python
sns.barplot(data= pltest, x='brand', y='price')
```
